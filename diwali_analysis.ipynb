# Gender wise data analysis

```python
# plotting a bar chart to understand the 
ax= sns.countplot(x='Gender',data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

# plotting a bar chart for gender and total amount

```python
sales = df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by='Amount',ascending=False)
ax =sns.barplot(x='Gender',y='Amount',data = sales)

for bar in ax.containers:
    ax.bar_label(bar)
```

* from above graph we can say that female spend more money as compare to the man. female has more purchasing power *

# now we are doing age wise analysis

```python
ax = sns.countplot(x= 'Age Group',hue = "Gender", data  = df )

for bar in ax.containers:
    ax.bar_label(bar)
```

```python
# analysis using total amount and age group


plt.figure(figsize = (10,5))
ax = sns.barplot(y = 'Age Group', x = 'Amount', hue = 'Gender',data =df,errorbar = None)

for bar in ax.containers:
    ax.bar_label(bar)
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending = False)

plt.figure(figsize = (10,5))
ax = sns.barplot(y = 'Age Group', x = 'Amount',data =sales_age,errorbar = None)

for bar in ax.containers:
    ax.bar_label(bar)
```

 # total number of order from states

```python
order = df.groupby(["State"],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=True)

plt.figure(figsize = (25,10))

ax = sns.barplot(y='Orders',x='State', data = order)

for bar in ax.containers:
    ax.bar_label(bar)
```

```python
# total number of sales by state

amnt = df.groupby(["State"],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

plt.figure(figsize = (25,10))

ax = sns.barplot(y='Amount',x='State', data = amnt)

for bar in ax.containers:
    ax.bar_label(bar)
```

# marital status

```python
marstatus = df.groupby(['Shaadi','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)


ax = sns.barplot(data = marstatus, x = 'Shaadi',y= 'Amount', hue='Gender')

for bar in ax.containers:
    ax.bar_label(bar)
```

```python
# sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount')
```

```python
sns.set(rc={'figure.figsize':(25,8)})

ax = sns.countplot(data = df, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
```

```python
# top 10 sold  products

fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
